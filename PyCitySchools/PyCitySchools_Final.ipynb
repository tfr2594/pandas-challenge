```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

schools_df = pd.DataFrame(school_data)
students_df = pd.DataFrame(student_data)

#I checked this works, and just the starter notebook to follow what I need to do

```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# calculate 
total_schools = school_data_complete['school_name'].nunique()
total_students = school_data_complete['Student ID'].count()
total_budget = schools_df['budget'].sum()

avg_math_score = school_data_complete['math_score'].mean()
avg_reading_score = school_data_complete['reading_score'].mean()

math_score_above70 = school_data_complete.loc[school_data_complete['math_score'] >= 70, :]
percent_passing_math = (math_score_above70.count()['Student ID'] / total_students) * 100

reading_score_above70 = school_data_complete.loc[school_data_complete['reading_score'] >= 70, :]
percent_passing_reading = (reading_score_above70.count()['Student ID'] / total_students) * 100

overall_passing = school_data_complete.loc[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70), :]
percent_passing_overall = (overall_passing.count()['Student ID'] / total_students) * 100

```

```python
#keeps repeatin, but just get one of them to look at it because I can't see anything
summary_df = pd.DataFrame({"Total Schools": [total_schools],
                              "Total Students": total_students,
                              "Total Budget": total_budget,
                              "Average Math Score": avg_math_score,
                              "Average reading Score": avg_reading_score,
                              "% Passing Math": percent_passing_math,
                              "% Passing Reading": percent_passing_reading,
                              "% Overall Passing": percent_passing_overall})
summary_df
```

```python
grouped_highschools = school_data_df.groupby(['school_name'])
```

```python
grouped_type_school = grouped_highschools['type'].min()
grouped_total_students = grouped_highschools['Student ID'].count()
grouped_total_budget = grouped_highschools['budget'].min()
grouped_budget_per_student = grouped_highschools['budget'].min() / grouped_highschools['Student ID'].count()
grouped_avg_math_score = grouped_highschools['math_score'].mean()
grouped_avg_reading_score = grouped_highschools['reading_score'].mean()

all_mathscores_over70 = school_data_complete[school_data_complete['math_score'] >= 70] 

grouped_highschools_math = all_mathscores_over70.groupby(['school_name']).count()['math_score']
Schools_math_score_above70 = (grouped_highschools_math / grouped_total_students) * 100

all_readingscores_over70 = school_data_complete[school_data_complete['reading_score'] >= 70]

grouped_highschools_reading = all_readingscores_over70.groupby(['school_name']).count()['reading_score']
Schools_reading_score_above70 = (grouped_highschools_reading / grouped_total_students) * 100

#rows_that turn up both = [(math scores above 70%) & (reading scores above 70%), :]
#percentage_overall_passing_highschools = (rows_that turn up both / grouped_total_students) * 100

passing_both_groups = school_data_complete[(school_data_complete['reading_score'] >= 70) &
                               (school_data_complete['math_score'] >= 70)].groupby('school_name')['Student ID'].count()
overall_percent_passing = passing_both_groups / grouped_total_students * 100 


```
